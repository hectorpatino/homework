#### Prompt

1. Cree dos archivos de tipo csv. En el primero debe colocar la información de 100 personas (clientes): id, cedula, nombre, apellido, número de cuenta.  En el segundo debe colocar transacciones de cada cliente. Por ejemplo (id= 3456, fecha = 29/01/2023, valor = -50_000). En total  mil transacciones.

```python
import csv
import random
from datetime import datetime, timedelta


import csv
import random

# Definir los nombres de los archivos CSV
clientes_csv = "clientes.csv"

# Definir el número de clientes
num_clientes = 100

# Crear la lista de encabezados para el archivo CSV
encabezados = ["id", "cedula", "nombre", "apellido", "numero_cuenta"]

# Crear la lista de clientes
clientes = []
for i in range(num_clientes):
    id_cliente = i + 1
    cedula = str(random.randint(10000000, 99999999))
    nombre = "Cliente" + str(id_cliente)
    apellido = "Apellido" + str(id_cliente)
    numero_cuenta = str(random.randint(1000000, 9999999))
    cliente = [id_cliente, cedula, nombre, apellido, numero_cuenta]
    clientes.append(cliente)

# Escribir la lista de clientes en el archivo CSV
with open(clientes_csv, mode="w", newline="") as archivo_csv:
    writer = csv.writer(archivo_csv, delimiter=",")
    writer.writerow(encabezados)
    writer.writerows(clientes)

print("Archivo CSV de clientes creado exitosamente.")

# Definir los nombres de los archivos CSV
transacciones_csv = "transacciones.csv"

# Definir el número de transacciones por cliente
num_transacciones = 10

# Obtener la lista de clientes del archivo CSV
clientes_csv = "clientes.csv"
clientes = []
with open(clientes_csv, mode="r") as archivo_csv:
    reader = csv.reader(archivo_csv, delimiter=",")
    next(reader) # Saltar la fila de encabezados
    for row in reader:
        clientes.append(row)

# Crear la lista de encabezados para el archivo CSV
encabezados = ["id_cliente", "fecha", "valor"]

# Crear la lista de transacciones
transacciones = []
for cliente in clientes:
    id_cliente = cliente[0]
    for i in range(num_transacciones):
        fecha = datetime.now() - timedelta(days=random.randint(1, 365))
        valor = random.randint(-1000000, 1000000)
        transaccion = [id_cliente, fecha.strftime("%d/%m/%Y"), valor]
        transacciones.append(transaccion)

# Escribir la lista de transacciones en el archivo CSV
with open(transacciones_csv, mode="w", newline="") as archivo_csv:
    writer = csv.writer(archivo_csv, delimiter=",")
    writer.writerow(encabezados)
    writer.writerows(transacciones)

print("Archivo CSV de transacciones creado exitosamente.")
```

```python
import pandas as pd

def leer_archivo_csv(nombre_archivo):
    # Leer el archivo CSV en un DataFrame de Pandas
    df = pd.read_csv(nombre_archivo)
    
    # Devolver el DataFrame
    return df
```

#### Prompt
Convierta el diccionario en un objeto json.

```python
def crear_diccionario(clientes_df, transacciones_df):
    # Agrupar las transacciones por cliente
    transacciones_por_cliente = transacciones_df.groupby("id_cliente").apply(lambda x: x.to_dict(orient="records")).to_dict()

    # Crear un diccionario con un registro por cliente
    clientes_dict = {}
    for _, row in clientes_df.iterrows():
        id_cliente = row["id"]
        nombre_cliente = row["nombre"] + " " + row["apellido"]
        cedula_cliente = row["cedula"]
        transacciones_cliente = transacciones_por_cliente.get(id_cliente, [])
        clientes_dict[id_cliente] = {"nombre": nombre_cliente, "cedula": cedula_cliente, "transacciones": transacciones_cliente}

    # Devolver el diccionario
    return clientes_dict

```

```python
clientes_dict = crear_diccionario(clientes_df, transacciones_df)
clientes_dict
```

```python
clientes_json = json.dumps(clientes_dict)
clientes_json
```

#### Prompt
Con base en todo lo anterior construya una clase, siguiendo las mejores prácticas que se verán en clase, que implemente toda la solución

```python
import pandas as pd
import json

class ManejoClientes:
    
    def __init__(self, clientes_file, transacciones_file):
        self.clientes_df = pd.read_csv(clientes_file)
        self.transacciones_df = pd.read_csv(transacciones_file)
        
    def crear_diccionario(self):
        clientes_dict = {}
        for idx, row in self.clientes_df.iterrows():
            id_cliente = row["id"]
            nombre = row["nombre"]
            cedula = row["cedula"]
            transacciones = []
            for _, t in self.transacciones_df[self.transacciones_df["id_cliente"]==id_cliente].iterrows():
                transacciones.append({
                    "fecha": t["fecha"],
                    "valor": t["valor"]
                })
            clientes_dict[id_cliente] = {
                "nombre": nombre,
                "cedula": cedula,
                "transacciones": transacciones
            }
        return clientes_dict
    
    def guardar_json(self, json_file):
        clientes_dict = self.crear_diccionario()
        clientes_json = json.dumps(clientes_dict)
        with open(json_file, "w") as f:
            json.dump(clientes_json, f)
```

```python
manejo_clientes = ManejoClientes("clientes.csv", "transacciones.csv")
manejo_clientes.guardar_json("clientes.json")
```

## Parte 1

```python
from typing import Union

def avg(cantidad_datos: Union[int, float], decimales: int = 2):
  values = []
  for value, idx in enumerate(range(cantidad_datos), 1):
    value_e = float(input(f"Ingrese el número {value}: "))
    values.append(value_e)
  suma = sum(values)
  registros = len(values)
  
  return print(f'{round(suma/registros, decimales)}')
```

```python
avg(cantidad_datos=3, decimales=3)
```

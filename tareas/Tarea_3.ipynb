```python
from datetime import datetime
```

```python
def calcular_edad(fecha_nacimiento):
    fecha_actual = datetime.now()
    fecha_nacimiento = datetime.strptime(fecha_nacimiento, '%d/%m/%Y')
    edad = fecha_actual.year - fecha_nacimiento.year
    if fecha_actual.month < fecha_nacimiento.month:
        edad -= 1
    elif fecha_actual.month == fecha_nacimiento.month and fecha_actual.day < fecha_nacimiento.day:
        edad -= 1
    print(f'La edad es {edad}')

fecha_nacimiento = '22/10/1990'
edad = calcular_edad(fecha_nacimiento)
```

```python
def centrar(texto):
    ancho = 80
    longitud = len(texto)
    espacios = (ancho - longitud) // 2
    print(' ' * espacios, texto)
centrar('Mi Nombre es Hector Patiño')
```

```python
def markdown_decorator(func):
    def wrapper(texto):
        return f"**_{func(texto)}_**"
    return wrapper

@markdown_decorator
def escribir_texto(texto):
    return texto

texto = 'Mi nombre es Hector Patiño'
resultado = escribir_texto(texto)
print(resultado)
```

```python
def multiplos(num1, num2):
  if num1 % num2 == 0:
      print(f"{num1} es múltiplo de {num2}")
  elif num2 % num1 == 0:
      print(f"{num2} es múltiplo de {num1}")
  else:
      print("Los números no son múltiplos entre sí")

multiplos(2 , 4)
```

```python
def temperatura_media(temperaturas):
    suma_temperaturas = sum(temperaturas)
    media_temperaturas = suma_temperaturas / len(temperaturas)
    media_temperaturas = round(media_temperaturas, 2)
    return media_temperaturas

temperaturas = [12.5, 13.1, 13.2, 13.8, 14.0, 14.5, 15.0, 15.2, 15.7, 16.1, 16.5, 16.9, 
                       17.3, 17.5, 17.7, 17.9, 18.0, 17.9, 17.7, 17.2, 16.8, 16.3, 15.8, 14.5]

media_temperaturas = temperatura_media(temperaturas)
print(f"La temperatura media en Bogotá hoy fue de {media_temperaturas}")
```

```python
def convertir(texto):
    resultado = ""   
    for letra in texto:   
        resultado += letra + " "    
    return resultado.rstrip()

resultado = convertir("Hola, tú")
print(resultado)
```

```python
def calcular_max_min(arreglo):    
    maximo = arreglo[0]
    minimo = arreglo[0]
    for elemento in arreglo:
        
        if elemento > maximo:
            maximo = elemento
        if elemento < minimo:
            minimo = elemento
    print("El valor máximo es:", maximo)
    print("El valor mínimo es:", minimo)
    return maximo, minimo
arreglo = [3, 5, 1, 9, 2, 7, 4]
maximo, minimo = calcular_max_min(arreglo)
```

```python
import math

def calcular_area_perimetro(radio):    
    area = math.pi * radio ** 2    
    perimetro = 2 * math.pi * radio
    print(f'Area= {area}')
    print(f'Perimetro = {perimetro}')
    return (area, perimetro)

resultado = calcular_area_perimetro(5)
```

```python
def login():
    username = input("Ingrese el nombre de usuario: ")
    password = input("Ingrese la contraseña: ")
    intentos = 0
    while username != 'usuario1' and password!='asdasd':
        if username == "usuario1" and password == "asdasd":
            return True, intentos
        else:
            intentos += 1
            print("Nombre de usuario o contraseña incorrectos. Inténtelo de nuevo.")
            username = input("Ingrese el nombre de usuario: ")
            password = input("Ingrese la contraseña: ")
    return False, intentos
```

```python
def factorial(num):
    if num == 0:
        return 1
    else:
        return num * factorial(num-1)

n = int(input("Ingresa un número: "))
print("El factorial de", n, "es", factorial(n))
```

```python
import functools

def media_geometrica(lista):
    producto = functools.reduce(lambda x, y: x * y, lista)
    n = len(lista)
    media = (producto)**(1/n)
    print(media)
    return media
```

```python
numeros = [2, 4, 6, 8, 10]
media = media_geometrica(numeros)
```
